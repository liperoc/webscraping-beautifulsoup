```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from sqlalchemy import create_engine
import psycopg2
URL = 'http://books.toscrape.com/catalogue/page-{page}.html'
```

```python
books=[]  # A list to store book

for page in range(1,51): # Loop through pages
    response = requests.get(URL.format(page=page))
    soup = BeautifulSoup(response.content, "html.parser")
    
    table = soup.find('ol', class_='row') 
    
    for row in table.findAll('article', class_='product_pod'): 
        book = {}
        book['Title'] = row.h3.a['title']
        book['Price'] = row.find('p', class_='price_color').text
        book['Rating'] = row.find('p', class_='star-rating')['class'][1]
        book['Availability'] = row.find('p', class_='instock availability').text.strip()
        books.append(book)

    print(book)
```

```python
df = pd.DataFrame(books, columns=['Title', 'Price', 'Rating', 'Availability']) # Creating DataFrame with Pandas
df
```

```python
df.to_csv('books_webscraping.csv') # Save DataFrame as CSV
```
